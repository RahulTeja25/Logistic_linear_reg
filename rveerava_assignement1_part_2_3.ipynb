```python
## Definging X and Y

X=df_1.loc[:,['x','carat']]
y=df_1['price']
```

```python
## spliting into train and test

indices = np.arange(X.shape[0])
train_ind = int(0.8 * X.shape[0])
np.random.shuffle(indices)
train_indices = indices[:train_ind]
test_indices = indices[train_ind:]


X_train, X_test = X.iloc[train_indices].values, X.iloc[test_indices].values
y_train, y_test = y.iloc[train_indices].values, y.iloc[test_indices].values
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
class LinearRegression():
    
    def __init__(self,X_train,y_train):
        self.X_train=X_train
        self.y_train=y_train
    def fit(self):
        r,c=self.X_train.shape
        self.bias=np.ones((r,1))
        self.X_train=np.concatenate((self.bias,self.X_train), axis=1)
        self.weights = np.dot(np.linalg.inv(np.dot(self.X_train.T,self.X_train)), np.dot(self.X_train.T,self.y_train))
    def predict(self,X_test,y_test):
        r,c=X_test.shape
        self.b=np.ones((r,1))
        X_test=np.concatenate((self.b,X_test), axis=1)
        self.y_hat=np.dot(X_test,self.weights)
        
        return self.y_hat
            
    def mse(self,X_test,y_test):
        r,c=X_test.shape
        X_test=np.concatenate((self.b,X_test), axis=1)
        self.y_pred=np.dot(X_test,self.weights)
        return np.mean((y_test-self.y_pred)**2)
```

```python
l=LinearRegression(X_train,y_train)
l.fit()
y_pred=l.predict(X_test,y_test)
print(l.weights)
print(y_test)
print(y_pred)

l.mse(X_test,y_test)
```

```python
## r2 score

y_mean=np.mean(y)
r2_score=np.sum((y_pred-y_mean)**2)/np.sum((y_test-y_mean)**2)
r2_score
```

Part 3   Ridge Regression

```python
class RidgeRegression():
    
    def __init__(self,X_train,y_train):
        self.X_train=X_train
        self.y_train=y_train
    def fit(self):
        r,c=self.X_train.shape
        self.bias=np.ones((r,1))
        self.X_train=np.concatenate((self.bias,self.X_train), axis=1)
        size=np.dot(self.X_train.T,self.X_train).shape[0]
        la=5
        self.weights = np.dot(np.linalg.inv(np.dot(self.X_train.T,self.X_train)+la*np.identity(size)), np.dot(self.X_train.T,self.y_train))
    def predict(self,X_test,y_test):
        r,c=X_test.shape
        self.b=np.ones((r,1))
        X_test=np.concatenate((self.b,X_test), axis=1)
        self.y_hat=np.dot(X_test,self.weights)
        
        return self.y_hat
            
    def mse(self,X_test,y_test):
        r,c=X_test.shape
        la=5
        X_test=np.concatenate((self.b,X_test), axis=1)
        self.y_pred=np.dot(X_test,self.weights)
        # lss=(1/X_train.shape[0])*((np.sum((y_test-self.y_pred)**2))+ (1/2)*la*np.dot(self.weights.T,self.weights))
        lss=np.mean((y_test-y_pred)**2)+(0.5/X_train.shape[0])*la*np.dot(self.weights.T,self.weights) ## Getting the current dimensions while using np.mean rather thsn using shape of X_train
        return lss
```

```python
r=RidgeRegression(X_train,y_train)
r.fit()
y_pred=r.predict(X_test,y_test)
print(r.mse(X_test,y_test))
print(r.weights)
```

```python
## r2 score

y_mean=np.mean(y)
r2_score=np.sum((y_pred-y_mean)**2)/np.sum((y_test-y_mean)**2)
r2_score
```
